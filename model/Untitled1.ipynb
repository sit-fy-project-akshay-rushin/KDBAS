```python
from random import uniform, randint
from math import floor, ceil
import numpy as np

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping, TensorBoard
```

```python
ll = 0.0001
ul = 3
sample_x = []
global fc_model
global graph
graph = tf.get_default_graph()
```

```python
def uniform_non_one(ll, ul):
    u = uniform(ll, ul)
    while u > 0.95 and u < 1.1:
        u = uniform(ll, ul)
    return u

def get_fakes(fakes):
    global ll, ul
    return np.sort(
        np.append(
            np.linspace(0,1, floor(fakes/2), endpoint=False), 
            np.linspace(3,1, ceil(fakes/2), endpoint=False) 
        ))
    
def get_train_data():
    global sample_x, ll, ul
    fakes = len(sample_x) + 2
    train_x = np.array(sample_x)
    train_y = np.ones(len(sample_x))
    fake_samples = get_fakes(fakes)
    
    for i in range(0, fakes):
        train_x = np.append(train_x, [train_x[randint(0, len(sample_x)-1)] * fake_samples[i]], axis=0)
        train_y = np.append(train_y, [0])
        
    return (train_x, train_y)
    
def get_model():
    input_len = len(sample_x[0])

    model = Sequential()
    # model.add(Dense(int((input_len+2)/2), input_shape=(input_len,)))
    model.add(Dense(int((input_len+2)*2), input_shape=(input_len,)))
    model.add(Dense(int((input_len+2)*4), activation="tanh"))
    model.add(Dense(1, activation="sigmoid"))
    model.summary()
    
    model.compile(optimizer=Adam(2e-4),
              loss='binary_crossentropy',
              metrics=['mse'])
    
    return model
```

```python
def train():
	global fc_model
	fc_model = get_model()
	x, y = get_train_data()
	print(fc_model.evaluate(np.array([[99, 357, 117, 99, 95, 83, 131, 119, 87, 110, 361, 185, 298, 261, 160, 193, 175, 169, 208, 215]]), np.array([1])))
	fc_model.fit(x, y, 
			  validation_split= 0.0, 
			  epochs=500,
			   callbacks=[ TensorBoard(),
                   EarlyStopping('loss', patience=50, restore_best_weights=True)
               ], 
			  batch_size=32)
	return 'trained'

def predict(test_x):
	prediction = fc_model.predict(test_x)
	print(prediction)
	return 'Prediction: ' + prediction
```
